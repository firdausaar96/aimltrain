```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('C:/Users/pc/Desktop/AiMLTraining/Day 13/netflix_data.csv')
df.head(3)
```

```python
sns.set_theme(style='white',palette='Set2')
sns.countplot(data=df,x='type')
plt.title('Number of Movies vs TV Shows')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.show()
```

```python
df=df.dropna(subset=['type','genre'])
```

```python
df['genre']
```

```python
df['country']
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df.head(30),x='type',hue='country',palette='Set2')
plt.title('Netflix Content type by Country')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.legend(title='Country',loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,x='type',hue='country',palette='Set2')
plt.title('Netflix Content type by Country')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.legend(title='Country',loc='upper left',frameon=True,bbox_to_anchor=(1,1),ncol=5)
plt.tight_layout()
plt.show()
```
