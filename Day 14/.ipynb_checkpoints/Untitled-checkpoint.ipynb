# Matplotlib.pyplot.subplots()

```python
# Subplot(): provides a way to plot multiple plot on single figure
            #function is the recommended and most common way to create a figure and a grid of subplots in a single call.It is more convenient than creating each subplot individually. 
# Given Number of rows and columns. It returns a tuple (fig,ax) giving a single figure with an array axes
```

```python
import matplotlib.pyplot as plt
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
```

```python
import numpy as np
```

```python
sales_data={'Electronics':45000,
            'Clothing':30000,
            'Groceries':25000,
            'Mobile':20000}
category=list(sales_data.keys())
actual_sale=list(sales_data.values())
print(category)
print(actual_sale)

fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
x=np.random.randn(10)
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(category,actual_sale,)
ax4.hist(np.random.random(1000))
```

# Subplot using index

```python
# Generate 100 evenly spaced numbers between 0 and 10 (inclusive)
our_line=np.linspace(0,10,100)
```

```python
product_prices={'Hair Oil':20,
            'Face Wash':30,
            'Lip Balm':10,
            'Face Cream':15}
```

```python
our_numbers=np.random.random(10)
```

```python
our_dots=np.random.randn(1000)
```

```python
#fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
fig,ax=plt.subplots(2,2,figsize=(10,5))
ax[0,0].plot(our_line,our_line/2)
ax[0,1].bar(list(product_prices.keys()),list(product_prices.values()),color='green')
ax[1,0].scatter(np.random.randn(1000),np.random.randn(1000))
ax[1,1].hist(our_dots,color='red')
#First row 0 & first column 0 [0,0]
#--------------------------------------------
ax[0,0].set_title('Line Plot Example')
ax[0,0].set_xlabel('X Axis')
ax[0,0].set_ylabel('Y Axis')
ax[0,0].grid(True,linestyle='--', alpha=0.7, color='grey')
fig.suptitle('MatplotLib Subplot example using 2 rows and 2 columns')
plt.tight_layout()
plt.savefig(('Subplot Example.jpeg'),dpi=300,bbox_inches='tight')
```
