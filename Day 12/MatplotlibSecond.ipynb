```python
import matplotlib.pyplot as plt
```

```python
products=['Laptop','Ipad','Headphones','Keyboard']
weekly_sales=[400,200,600,150]
plt.bar(products,weekly_sales)
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

```python
products=['Laptop','Ipad','Headphones','Keyboard']
weekly_sales=[400,200,600,150]
plt.figure(figsize=(6,3))
bar_colors=['Blue','Red','Green','Yellow']
plt.bar(products,weekly_sales,color=bar_colors, width=0.5, edgecolor='Black')
plt.title('Weekly Product Sales')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

# We have following dictionary 
    ## product_sales={'Laptop':750,'SmartPhones':650,'SmartWatch':400,'Tablet':500,'AC':400}

    ## Represent these data using bar chart x-label is ProductName and y-Label is sales value

```python
product_sales={'Laptop':750,'SmartPhones':650,'SmartWatch':400,'Tablet':500,'AC':400}
products=list(product_sales.keys())
sales=list(product_sales.values())
plt.bar(products,sales,color='Green', width=0.5, edgecolor='Black')
plt.title('Product Sales')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

# Pie Chart

```python
products=['Laptop','Ipad','Headphones','Keyboard']
sales=[400,200,600,150]
plt.figure(figsize=(5,5))
plt.pie(sales, labels=products)
plt.title('Product Weekly Sales Report')
plt.show()
```

```python
plt.figure(figsize=(5,5))
product_sales={'Laptop':750,'SmartPhones':650,'SmartWatch':400,'Tablet':500,'AC':400}
products=list(product_sales.keys())
sales=list(product_sales.values())
pie_color=['blue','red','green','yellow','skyblue']
plt.pie(sales,labels=products,color=pie_color,shadow=True,startang=45)
plt.title('Product Weekly Sales Report')
plt.show()
```

# Scatter Plots

```python
product_sales={'Laptop':750,'SmartPhones':650,'SmartWatch':400,'Tablet':500,'AC':400}
products=list(product_sales.keys())
sales=list(product_sales.values())
plt.scatter(products, sales)
plt.title('Basic Product and Sales Representation')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
players=['MSD','Rohit','Virat','Gill','Firdaus']
match1_score=[45,67,56,78,89]
match2_score=[55,78,40,80,97]
plt.figure(figsize=(8,5))
plt.scatter(players,match1_score, color='red',label='Match One', s=100)
plt.scatter(players,match2_score, color='green',label='Match Two', s=100, marker='x')
plt.title('Comparison of Players in Two Matches')
plt.xlabel('Players')
plt.ylabel('Score')
plt.grid(True,linestyle='--', alpha=0.5)
plt.legend(loc='best')
plt.show()
```

```python
## Write a comparison that show average calories burned for each workout type
```

```python
df=pd.read_csv('Final_data.csv')
avg_cal=df.groupby('Workout_Type')['Calories_Burned'].mean()
selected_index=avg_cal.index
selected_value=avg_cal.values
plt.figure(figsize=(8,5))
workout_type=df['Workout_Type']
plt.scatter(selected_index,selected_value, color='green',label='Calories Burned', s=100, marker='o')
plt.title('Comparison of Average Calories Burned for each Workout Type')
plt.xlabel('Workout Type')
plt.ylabel('Calories Burned')
plt.grid(True,linestyle='--', alpha=0.5)
#plt.legend(loc='best')
plt.show()
```

# Create a bar chart using following data
    ## we following libraries 
        ## python_lib=['Numpy','Pandas','Matplotlib','Scikit-learn']
    ## Liked by users as follows
        ## liked_by=[5000, 7000,9000, 4000]
    ## Create a vertical bar graph where each bar represents the value of its respective likedby users

```python
python_lib=['Numpy','Pandas','Matplotlib','Scikit-learn']
liked_by=[5000, 7000,9000, 4000]
plt.figure(figsize=(6,3))
bar_colors=['Blue','Red','Green','Yellow']
plt.bar(python_lib,liked_by,color=bar_colors, width=0.5, edgecolor='Black')
plt.title('Most Likeable Python Library')
plt.xlabel('Python Library')
plt.ylabel('Likes')
plt.show()
```

```python
plt.figure(figsize=(4,3))
bar_colors=['Blue','Red','Green','Yellow']
plt.barh(python_lib,liked_by,color=bar_colors)
plt.title('Most Likeable Python Library')
plt.show()
```
