```python
x=[1,2,3,4,5]
```

```python
y=[10,20,15,30,45]
```

# Basic of Matplotlib chart
    ## pyplot: interface for creating and managing plot
    ## Figure: The complete window to create plot
    ## Axes: The area where data is plotted
    ## Axis: x line and y line to define scale and ticks
    ## plot: the graph of data

```python
plt.figure(figsize=(8,6))
```

```python
plt.plot(x,y,marker='o',linestyle='-')
```

```python
plt.title('First Line Chart')
```

```python
plt.xlabel('X Label')
```

```python
plt.ylabel('Y Label')
```

```python
plt.grid(True)
```

# First Example to create a line chart from our data
    ## Student.xlsx for fetching the data
    ## Generate Line Chart based on Student Name and Student Marks

```python
import pandas as pd
```

```python
df=pd.read_excel('C:/Users/pc/Desktop/AiMLTraining/Day 10/students.xlsx')
```

```python
names=df['Name']
marks=df['Marks']
```

```python
plt.plot(names,marks,marker='s', color='green')
plt.title('Student Mark Line Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(names,marks,marker='o', color='orange',linestyle='-')
plt.title('Student Progress Chart')
plt.xlabel('Name')
plt.ylabel('Marks')
plt.show()
```

```python
print('Dashed Line')
plt.plot(names,marks,marker='o', color='orange',linestyle='--')
plt.show()
```

```python
print('Dotted Line')
plt.plot(names,marks,marker='o', color='orange',linestyle=':')
plt.show()
```

```python
print('Dashed dotted Line')
plt.plot(names,marks,marker='o', color='orange',linestyle='-.')
plt.show()
```
