# KNN Algorithm


```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
x=np.array([
    [1,3500],
    [2,3200],
    [3,3000],
    [4,2500], 
    [5,2300],
    [6,2000],
    [7,2200],
    [2,3700], 
    ])
y=['unfit','unfit','unfit','Fit','Fit','Fit','Fit','unfit']
```

```python
#create the model
model=KNeighborsClassifier(n_neighbors=3)
```

```python
#Train the model
model.fit(x,y)
```

```python
#predict the data
new_person=[[4.5,2600]]
pred=model.predict(new_person)
print('Predicted Result: ',pred[0])
```

```python
#Prediction from taking the values from user
exercise_hours=float(input('Enter Number of Exercise Hours'))
cal_intake=float(input('How many calories did you consumed'))
pred_fu=model.predict([[exercise_hours,cal_intake]])
print(f'Exercise Hours: {exercise_hours}\nCalorie Intake: {cal_intake}\n. You are {pred_fu[0]}')
```

```python
import pandas as pd
```

```python
df=pd.read_csv('fruit_data.csv')
df
```

```python
#Split the data
x=df[['Weight','Size']]
y=df['Fruit']
```

```python
#Model Creation
model=KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(x,y)
```

```python
#Prediction from taking the values from user
fruit_weight=float(input('Enter weight of the fruit in gram'))
fruit_size=float(input('Enter size of the fruit in cm'))
pred_fruit=model.predict([[fruit_weight,fruit_size]])
print(f'Fruit Weight: {fruit_weight}\nFruit Size: {fruit_size}\n.The fruit is {pred_fruit[0]}')
```

```python
fruit_weight=float(input('Enter weight of the fruit in gram'))
fruit_size=float(input('Enter size of the fruit in cm'))
our_fruit=pd.DataFrame([[fruit_weight,fruit_size]],columns=['Weight','Size'])
pred_fruit=model.predict(our_fruit)
print(f'Fruit Weight: {fruit_weight}\nFruit Size: {fruit_size}\n.The fruit is {pred_fruit[0]}')
```

```python
# Downlaod student_grade.csv from GitHub.com
# Predict a student's grade level (A, B, or C)
# based on hours studied and assignments completed
# using the K-Nearest Neighbors algorithm
```

```python
data=pd.read_csv('student_grade.csv')
data
```

```python
#Split the data
x=data[['Hours_Studied','Assignments_Completed']]
y=data['Grade']
```

```python
model=KNeighborsClassifier(n_neighbors=3)
model.fit(x,y)
```

```python
hours_studied=float(input('Enter How many hours you have studied'))
asgmt_cmpleted=float(input('Enter How many assignment you have been completed'))
our_grade=pd.DataFrame([[hours_studied,asgmt_cmpleted]],columns=['Hours_Studied','Assignments_Completed'])
pred_grade=model.predict(our_grade)
print(f'\nHours Studied: {hours_studied}\nAssignment Completed: {asgmt_cmpleted}\nYour Grade is {pred_grade[0]}')
```
