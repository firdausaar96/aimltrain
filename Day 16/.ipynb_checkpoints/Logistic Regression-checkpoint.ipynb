```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df=pd.read_csv('pass_fail.csv')
df.head()
```

```python
x=df[['Hours']]
y=df['Result']
```

```python
model=LogisticRegression()
model.fit(x,y)
```

```python
pred_35=model.predict([[3.5]])
print('Predicted Result for 3.5 hours: ',pred_35[0])
```

```python
hours_study=float(input('Numbers of Hours Studied'))
pred_value=model.predict([[hours_study]])
print(f'After {hours_study} hours study you are likely to {pred_value[0]} ')
```

```python
#Import all required libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
```

```python
#collecting data
df=pd.read_csv('customer_purchase.csv')
df.head()
```

```python
le=LabelEncoder()
df['Purchased']=le.fit_transform(df['Purchased'])
x=df[['Age','Salary']]
y=df['Purchased']
```

```python
df
```

```python
#Create the model
model=LogisticRegression()

#Trained the model
model.fit(x,y)
```

```python
new_cust_age=int(input('Enter Age' ))
new_cust_sal=int(input('Enter Salary' ))
new_cust=[[new_cust_age,new_cust_sal]]
prediction=model.predict(new_cust)

if (prediction):
    print( 'Purchased')
else:
    print('Not Purchased')
```

# Logistic Regression - Spam/Not Spam Exercise

```python
data=pd.read_csv('spam_dataset.csv')
data.head()
```

```python
#Cleaning the data
print(data.isnull().sum())
```

```python
cv = CountVectorizer()
x=cv.fit_transform(data['message_content'])
y=data['is_spam']
```

```python
#Create the model
model=LogisticRegression()

#Trained the model
model.fit(x,y)
```

```python

new_massage_content=input('Enter your Email: \n' )
new_massage_vector=cv.transform([new_massage_content])
prediction=model.predict(new_massage_vector)

if (prediction):
    print( '--> SPAM <--')
else:
    print('--> Not SPAM <--')
```
