```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('sales_data.csv')
data.head()
```

```python
x=data[['Advertising_Spend']]
y=data['Sales_Revenue']
```

```python
model=LinearRegression()
model.fit(x,y)
print('Training Done')
```

```python
new_spend=float(input('Enter New Spend to Predict Sales Revanue'))
predicted_revenue=model.predict([[new_spend]])
print(f'New Spend {new_spend} Predicted Revanue {predicted_revenue.round(2)} ')
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('house_price.csv')
df.head()
```

```python
size=df[['Size_sqft']]
price=df['Price']
```

```python
model=LinearRegression()
model.fit(size,price)
```

```python
new_size=float(input('Enter New Size to Predict the House Price '))
predicted_price=model.predict([[new_size]])
print(f'New Price for the {new_size} sqft is RM{predicted_price.round(2)} ')
```
