```python
import matplotlib.pyplot as plt
```

```python
months=['Jan','Feb','March','April']
sales=[2000,3000,2500,3600]
```

# ------------------------------------------------

```python
# plt.plot(x,y,'color name',linestyle='linestyle',width=value, marker='marker',label='label name'
```

```python
plt.plot(months,sales, color='hotpink', linestyle='--',linewidth=3, marker='p',label='Monthly sales')
plt.xlabel('Months')
plt.ylabel('Sales Per Month')
plt.title('Monthly Sales Report')
plt.xlim(0,4)
plt.ylim(1800,4000)
plt.grid(True)
plt.legend(loc='best')
plt.show()
```

```python
import numpy as np
```

```python
x=np.linspace(0,1,100) # Generate 100 numbers between 0 and 1
plt.plot(x,x,label='Linear')
plt.plot(x,x**2, label='Quadratic')
plt.plot(x,x**3, label='Cubic')
plt.xlabel('X lines')
plt.ylabel('Y Lines')
plt.title('Simple Lines Plot')
plt.legend()
plt.show()
```

```python
# Bar Plot Example
# plt.bar(x,y,color='',edgecolor='')
```

```python
categories=['Electronics','Mobile','Book','Groceries','Home Decoration']
sales=[50000,60000,30000,90000,40000]
bar_colors=['hotpink','turquoise','magenta','red','yellow']
plt.figure(figsize=(8,5))
plt.bar(categories,sales,color=bar_colors, edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Sales Amount (MYR)')
plt.grid(axis='y',linestyle='--',color='grey',alpha=0.2)
plt.tight_layout()
plt.title('Categories wise Sales Report')
plt.show()
```

```python
categories=['Electronics','Mobile','Book','Groceries','Home Decoration']
sales=[50000,60000,30000,90000,40000]
plt.figure(figsize=(8,5))
plt.bar(categories,sales,color='blue', edgecolor='black',label='Size Info')
plt.xlabel('Categories')
plt.ylabel('Sales Amount (MYR)')
plt.grid(axis='y',linestyle='--',color='grey',alpha=0.4)
plt.xticks (rotation=90, fontsize=12, color='magenta')
plt.yticks ([20000,40000,60000,80000,100000],['20k','40k','60k','80k','100k'],fontsize=12, color='green')
plt.tight_layout()
plt.legend()
plt.title('Categories wise Sales Report')
plt.show()
```

# Scatter Plot

```python
# plt.scatter(x,y,s='',c='color name',marker='markername',alpha=transp,edgecolor='colorname',label='label of legend'
```

```python
height_cm=[150,155,160,165,170,175,180,185]
weight_kg=[50,53,59,63,66,68,72,78]
plt.figure(figsize=(8,8))
plt.scatter(height_cm,weight_kg,s=100,c='green',marker='*',alpha=0.9,edgecolor='black',label='weight height comparison')
plt.title('Scatter Plot Height vs weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(color='lightblue',alpha=0.5,linestyle='--')
plt.show()
```

```python
height_cm=[150,155,160,165,170,175,180,185]
weight_kg=[50,53,59,63,66,68,72,78]
age_year=[20,22,25,28,30,35,38,40]
```

```python
plt.figure(figsize=(8,8))
```

```python
scatter=plt.scatter(height_cm,weight_kg,c=age_year,cmap='rainbow',s=120,edgecolor='black')
plt.colorbar(label='Age (years)')
plt.title('Scatter Plot Height vs weight (colored by age)')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(color='lightblue',alpha=0.5,linestyle='--')
plt.show()
```

```python
# Pie Plot Example
    #plt.pie(x,labels='',colors='',autopct='',startangle='',shadow=True/False, explode=,radius=

##autopct - for pecentage
##startangle-whih angle want to start
```

```python
dept=['HR','Web-Development','App-Development','Agile-Scrum','Testing']
expenses=[12000,25000,15000,18000,20000]
```

```python
plt.figure(figsize=(8,8))
```

```python
plt.pie(expenses,labels=dept,autopct='%1.2f%%',startangle=0,explode=(0.09,0.05,0.05,0.05,0.05),shadow=True)
plt.title('Department-wise Expenses Report')
plt.axis('equal')     # Ensures the pie chart is drawn as a circle
plt.show()
```

```python
plt.pie(expenses,labels=dept,colors=plt.cm.magma([0.1,0.3,0.5,0.7,0.9]),autopct='%1.2f%%',startangle=0,explode=(0.09,0.05,0.05,0.05,0.05),shadow=True)
plt.title('Department-wise Expenses Report')
plt.axis('equal')     # Ensures the pie chart is drawn as a circle
plt.show()
```

```python
# Create pie chart using following data
## sales_data={'Mobile':45000,'Clothing':30000,'Groceries':35000,'Home Decor':20000}
## show percentage with 2 decimal places
## use colormap to give the different color to your pie
```

```python
sales_data={'Mobile':45000,'Clothing':30000,'Groceries':35000,'Home Decor':20000}
sales=list(sales_data.keys())
data=list(sales_data.values())
plt.pie(data,labels=sales,colors=plt.cm.rainbow([0.1,0.4,0.8,1.2]),textprops={'color':'white','fontsize':10,'fontweight':'bold'},autopct='%1.2f%%',startangle=0,shadow=True)
plt.title('Sales Report')
plt.savefig('SalesReport.pdf',dpi=300,bbox_inches='tight')
plt.show()
```

```python
marks=[45,60,80,55,62,90,89,50,56,60,80,55,62,90,89,58,90,56,67,83]
plt.style.use('ggplot')
plt.figure(figsize=(8,5))
plt.hist(marks,bins=6,color='skyblue',edgecolor='darkblue',alpha=0.8)
plt.title('Histogram for Student Marks')
plt.xlabel('Marks Range')
plt.ylabel('Number of Students')
plt.grid(color='grey',alpha=0.2,linestyle='--')
plt.show()
```
