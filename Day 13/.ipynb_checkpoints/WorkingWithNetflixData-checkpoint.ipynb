```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('netflix_data.csv')
print(df.head())
```

```python
df=df.dropna(subset=['release_year']) #remove row if they contain missing values
```

```python
df['release_year']=df['release_year'].astype(int)
plt.hist(df['release_year'],bins=100,color='green',edgecolor='darkblue',alpha=0.8)
plt.title('Netflix released Title by Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(color='grey',alpha=0.2,linestyle='--')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
df=pd.read_csv('netflix_data.csv')
df=df.dropna(subset=['release_year'])
df['release_year']=df['release_year'].astype(int)
plt.hist(df['release_year'],bins=100,color='green',edgecolor='darkblue',alpha=0.8)
plt.title('Netflix released Title by Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(color='grey',alpha=0.2,linestyle='--')
plt.savefig('Netflix Bar.pdf',dpi=500,bbox_inches='tight')
plt.show()
```

```python
df=df.dropna(subset=['release_year','type'])
```

```python
df['release_year']=df['release_year'].astype(int)
```

```python
movies=df[df['type']=='Movie']
show=df[df['type']=='TV Show']
```

```python
plt.figure(figsize=(10,5))
plt.hist([movies['release_year'],show['release_year']],bins=30,color=['green','red'],edgecolor='black',alpha=0.8,label=['Movies','TV Show'])
plt.title('Netflix released by Year and Type')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles Released')
plt.grid(color='grey',alpha=0.2,linestyle='--')
plt.legend(title='Movies/Shows')
plt.savefig('Netflix showmovies.jpg',dpi=500,bbox_inches='tight')
plt.show()
```

```python
# Use netflix_data.csv dataset
    #print how many Movies and Tv Shows are available on Netflix
    #Display Movies and TV Shows proportion in a pie chart
    #save file as pdf
```

```python
type_counts=df['type'].value_counts()

plt.figure(figsize=(10,5))
plt.pie(type_counts,
        labels=type_counts.index,
        colors=['green','blue'],
        explode=(0.05,0.00),
        autopct='%1.2f%%',
        startangle=0)
plt.title('Netflix Movies and TV Shows Distribution')
plt.legend(title='Movies and Shows')
plt.axis('equal')     # Ensures the pie chart is drawn as a circle
plt.savefig('Netflix Pie Plot.png',dpi=500,bbox_inches='tight')
plt.show()
```
